```python
# 设置在jupyter lab 中显示，
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NotebookType = NotebookType.JUPYTER_LAB

from pyecharts import charts
from pyecharts import options as opts     # 设置
from pyecharts.globals import ThemeType   # 主题
```

```python
# 全局配置项可通过set_global_opts 方法设置
bar = (
    charts.Bar(init_opts=opts.InitOpts(theme=ThemeType.LIGHT))
    .add_xaxis(['衬衫', '羊毛衫', '裤子', '袜子', '鞋'])
    .add_yaxis('商家A', [12, 15, 17, 7, 43])
    .add_yaxis('商家B', [14, 12, 20, 10, 23])
    # 全局配置
    .set_global_opts(
        # 标题配置项，此外还可以配置标题的字体、位置等
        title_opts=opts.TitleOpts(
            title='主标题', title_link='https://cn.bing.com',
            subtitle='副标题', subtitle_link=''    # 标题跳转链接
        ),
        # 工具箱置顶
        toolbox_opts=opts.ToolboxOpts(),
        # 工具箱配置
#         toolbox_opts=opts.ToolBoxFeatureOpts()
    )
    
    # 系列配置项
    .set_series_opts()
    
)

bar.render()
# bar.load_javascript()
# bar.render_notebook()
```
